```python
# 원본 데이터 백업
df_preprocessed = df_part.copy()

df_preprocessed['cats'] = list(map(np.array, cat_ohe))
df_preprocessed['campaigns'] = list(map(np.array, camp_ohe))


# 기본 개념

# list(map(np.array, cat_ohe))는 cat_ohe 리스트에 있는 각 요소를 np.array 함수를 사용하여 배열로 변환한 후, 그 결과를 리스트로 반환하는 과정입니다.

# 단계별 설명

    # map 함수:
# map 함수는 두 개의 인수를 받습니다: 함수와 이터러블(반복 가능한 객체).
# 여기서 함수는 np.array이고, 이터러블은 cat_ohe입니다.
# map 함수는 이터러블의 각 요소에 대해 주어진 함수를 적용하고, 그 결과를 맵 객체로 반환합니다.

    # np.array 함수:
# np.array는 리스트나 튜플과 같은 시퀀스를 NumPy 배열로 변환합니다.
# 이 경우, cat_ohe 리스트의 각 요소를 NumPy 배열로 변환합니다.

    # list 함수:
# map 함수는 맵 객체를 반환하기 때문에, 이를 리스트로 변환하기 위해 list 함수를 사용합니다.
# 최종 결과는 각 요소가 NumPy 배열인 리스트가 됩니다.





# 데이터 프레임 입장에서 원핫인코딩은 원핫인코딩 칼럼이 옆쪽으로 굉장히 많이 붙는건데, 
# 데이터만 나와있는 데이터 프레임에 인코딩 칼럼이 여러 개 있는 건 비효율적이기 때문에 
# 하나의 칼럼에 원핫인코딩된 것을 리스트 형태로 각각의 해당하는 행에 추가





# 원핫 인코딩(One-Hot Encoding)은 각 카테고리 값을 이진 값으로 표현하는 방법입니다. 
# 예를 들어, 'Cat1', 'Cat2', 'Cat3'라는 세 가지 카테고리가 있다면, 
# 이 카테고리들은 [1, 0, 0], [0, 1, 0], [0, 0, 1]과 같이 각각의 컬럼으로 변환됩니다.
# 이러한 변환은 카테고리 수가 많아지면 데이터프레임에 매우 많은 컬럼을 추가하게 됩니다.

# 이러한 방식은 데이터 프레임을 복잡하고 커지게 만들며, 메모리 사용량도 증가시킵니다.
# 따라서, 각 행에 해당하는 원핫 인코딩된 데이터를 하나의 리스트로 묶어 하나의 컬럼에 저장하는 것이 더 효율적일 수 있습니다.

# 위의 코드에서는 'cat_ohe'와 'camp_ohe'라는 원핫 인코딩된 데이터를 리스트 형태로 변환한 후,
# 각각 'cats'와 'campaigns'라는 새로운 컬럼에 저장합니다.
# 이렇게 하면, 원핫 인코딩된 여러 개의 컬럼이 각 행마다 하나의 리스트로 묶여서 저장되므로,
# 데이터 프레임이 훨씬 간결하고 관리하기 쉬워집니다.
# 예를 들어, 각 행의 'cats' 컬럼에는 해당 카테고리의 원핫 인코딩된 배열이 들어가게 됩니다.

# 이렇게 하면 각 행의 원핫 인코딩된 데이터를 하나의 리스트로 관리할 수 있어,
# 데이터 프레임의 크기를 줄이고 가독성을 높일 수 있습니다.





    # 요약
# 이 코드는 cat_ohe 리스트의 각 요소를 np.array를 사용하여 NumPy 배열로 변환하고, 그 결과를 리스트로 저장하는 과정입니다. 
# 이를 통해 데이터 전처리나 모델 입력 준비와 같은 작업을 효율적으로 수행할 수 있습니다.
```

## Logistic Regression 을 활용한 마케팅 예산 분배 최적화
* conversion 을 예측하는 모델을 만들고, 전환을 예측했을 때의 logistic regression 의 각 캠페인에 대한 학습된 가중치 (딥러닝에서는 weight, 회귀분석에서는 coefficient 라고 한다) 를 가져온다
* Logistic Regression 의 Weight  =>  attribution weights

```python
# series : cats, campaigns
#  One Hot Encoding -> Many Hot Encoding

def pairwise_max(series):
    return np.max(series.tolist(), axis=0).tolist()




# 마케팅 예산 분배 최적화
# = (어떤 캠페인에 투자를 많이 해야 많이 살까/어떤 캠페인에 돈을 덜 투자해야 비용을 줄일까)
#어떤 켐페인에 돈을 지불했을 때(어떤 캠페인에 투자를 많이 해야 많이 살까/어떤 캠페인에 돈을 덜 투자해야 비용을 줄일까) 유저가 살지 안 살지가 중요함 = 캠페인을 봤다 안봤다?



# 기본 개념

# OneHot 인코딩: 각 카테고리 값을 이진 벡터로 변환하는 방법이에요. 
# 예를 들어, 카테고리가 'A', 'B', 'C'가 있다면, 'A'는 [1, 0, 0], 'B'는 [0, 1, 0], 'C'는 [0, 0, 1]로 변환돼요.
# Many Hot 인코딩: 여러 개의 OneHot 벡터를 하나로 합치는 과정이에요. 
# 예를 들어, 여러 번의 노출이 있을 때, 각 노출에서 카테고리가 어떻게 나타나는지 하나의 벡터로 결합하는 거예요.
```

```python
aggregations = {
    "campaigns" : pairwise_max,
    "cats" : pairwise_max,
    "click" : 'sum',
    'cost' : 'sum',
    'conversion' : 'max'
}
```

```python
df_agg = df_preprocessed.groupby(['jid']).agg(aggregations)
df_agg.head()

# conversion을 통해 돈이 얼마가 들어갔고, 
# 전환으로 이어졌는지에 대해 판단해 어떠한 사용자한테 효과적이지 않은 광고 비용을 지불했는지를 볼 수 있음
```

```python
df_agg['features'] = df_agg[['campaigns', 'cats', 'click', 'cost']].values.tolist()
df_agg['features'].head()
```

```python
X = np.stack(df_agg['features'].map(lambda x : np.hstack(x)).values)
y = df_agg['conversion'].values
```

```python
X.shape, y.shape
```

376개의 Feature 중 처음 50

# 데이터 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=100000)
log_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_train_pred = log_reg.predict(X_train)
y_test_pred  = log_reg.predict(X_test)

print("훈련 데이터 정확도 : {:.3f}".format(accuracy_score(y_train, y_train_pred)))
print("테스트 데이터 정확도 : {:.3f}".format(accuracy_score(y_test, y_test_pred)))
```

```python
n_campaigns = 50

weights = log_reg.coef_[0, :n_campaigns]
weights.shape

# log_reg.coef_는 파이썬에서 LogisticRegression 모델을 사용하여 학습을 마친 후에, 
# 각 특징(feature)에 대한 가중치(coefficient)를 확인할 때 사용하는 속성입니다.
# 이 가중치들은 모델이 예측을 할 때 각 특징이 얼마나 중요한지를 나타냅니다.
```

# 마케팅 예산 분배 시뮬레이션
* 예산을 지정한다.
* 예산을 캠페인의 기여도에 맞게 분배한다.
* 과거 유저의 전환 내역을 Replay 한다.
    * 캠페인이 노출되었다면 각 캠페인의 예산을 차감하고
    * 전환이 실제 되었다면 전환에 + 1을 해준다
    * 캠페인이 과거에 노출되었는데, 해당 캠페인에 대한 예산은 다 쓴 상태라면 blacklist 에 넣어준다.
* 계산은 conversion - blacklist 로 한다.

```python
def get_campaign_id(x_journey_step, n_campaigns=50):
  '''
    x_journey_step : 각 행( 유저 한명 한명의 데이터 )의 캠페인
  '''
  return np.argmax(x_journey_step[0:n_campaigns]) # 한 사람한테 노출된 캠페인의 인덱스(0 ~ 49)

def simulate_budget_roi(df, budget_total, attribution):
  # 기여도에 따른 예신 분배 후 캠페인 별 시뮬레이션 진행
  # 예산이 100만원 일 때 기여도가 0.5이면 50만원을 가질 수 있도록 한다.

  # 각 캠페인 별 예산 구하기(올림 처리)
  # budgets = np.ceil(attribution * budget_total)

  # 방어코드 작성 스타일
  budgets = np.ceil(attribution * (budget_total / np.sum(attribution)))

  # 예산을 다 써서 시뮬레이션을 진행할 수 없는 경우
  blacklist = set()

  # 전환이 일어난 경우
  conversions = set()

  # 데이터 세트를 모두 돌면서 시뮬레이션 시작
  for i in range(df.shape[0]):

    # 해당 이벤트에서 노출된 캠페인 id를 가져오기. 0번인지, 1번인지, 35번인지..
    campaign_id = get_campaign_id(df.loc[i]['campaigns'])

    # 해당 jid를 가져온다.
    jid = df.loc[i]['jid']

    # jid가 blacklist에 없다 -> 시뮬레이션 수행이 가능(노출이 가능)
    if jid not in blacklist:

      # 캠페인에 대한 예산이 있으면
      if budgets[campaign_id] >= 1:

        # 예산에서 1원을 차감
        budgets[campaign_id] -= 1

        # 전환 여부 판단
        if df.loc[i]['conversion'] == 1:
          # 전환이 되었으면 전환에 추가
          conversions.add(jid)

      else:
        # 예산이 없으면
        # 이 사람(jid)에 대한 시뮬레이션은 멈춘다.
        blacklist.add(jid)

    # 확실하게 예산 내에서 전환된 사람의 수만 구하기
    # 과거에는 전환이 됐지만, 예산이 부족해 광고를 못 본 사람은 전환이 안 될것이다 라고 생각!
  return len(conversions.difference(blacklist))
```
