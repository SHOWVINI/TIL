```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import matplotlib as mpl


# 맥북은 설정하는 방법이 달라 보류

count = 0
for font in fm.fontManager.ttflist:
    # print(font.fname)
    count += 1
    if '/Users/evan/Library/Fonts/Nanum' in font.fname and count <= 100:
        print(count, font.name, font.fname)

fontpath = '/Users/evan/Library/Fonts/NanumMyeongjo.ttf'
font = fm.FontProperties(fname=fontpath, size = 24)
plt.rc('font', family='NanumMyeongjo') 
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 
mpl.font_manager._rebuild()
fm._rebuild()

plt.plot([1, 2, 3, 4, 5])
plt.title("테스트", fontproperties=font)
plt.show()
```

## `rcParams` 객체

rcParams 객체는 Matplotlib의 런타임 구성(runtime configuration) 설정을 관리하는 딕셔너리 객체입니다. 
rcParams를 통해 그래프의 기본 스타일과 속성을 설정하거나 변경할 수 있습니다. 
이 객체를 사용하여 폰트 크기, 색상, 선 스타일, Figure 크기 등 다양한 Matplotlib 설정을 조정할 수 있습니다.

## 여러 개의 서브플롯 출력하기

```python
fig, axs = plt.subplots(2)
# ax로 쓰나 axs로 쓰나 상관 없긴 함. 단지 가독성 위해 ㅋㅋ.

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_yscale('log')

fig.show()

# 이 코드는 Matplotlib를 사용하여 두 개의 서브플롯을 생성하는 예제입니다. 
# 첫 번째 서브플롯에는 산점도를 그리고, 두 번째 서브플롯에는 히스토그램을 그리며 y축을 로그 스케일로 설정합니다. 
# 각 부분의 의미와 기능을 설명하면 다음과 같습니다:

    # fig, axs = plt.subplots(2):

# Figure와 두 개의 서브플롯(AxesSubplot)을 생성합니다.
# 2는 2개의 서브플롯을 세로로 쌓아 배치함을 의미합니다. 반환되는 axs는 두 개의 서브플롯을 포함하는 배열입니다.

    # axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1):

# 첫 번째 서브플롯에 산점도를 그립니다.
# ns_book7['도서권수']는 x축 데이터, ns_book7['대출건수']는 y축 데이터로 사용됩니다.
# alpha=0.1은 점의 투명도를 설정하여 데이터가 겹칠 때도 볼 수 있도록 합니다.

    # axs[1].hist(ns_book7['대출건수'], bins=100):

# 두 번째 서브플롯에 히스토그램을 그립니다.
# ns_book7['대출건수']는 히스토그램의 데이터로 사용됩니다.
# bins=100은 히스토그램의 빈(bin) 개수를 100개로 설정합니다.

    # axs[1].set_yscale('log'):

# 두 번째 서브플롯의 y축을 로그 스케일로 설정합니다.
# 이는 데이터가 큰 범위를 차지할 때 유용합니다.

    # y축을 로그 스케일로 설정한다는 것은 y축의 눈금이 로그(대수) 값을 기반으로 설정된다는 의미입니다. 
    # 일반적인 선형 스케일과 달리 로그 스케일에서는 각 눈금이 고정된 배율(예: 10배)로 증가합니다. 
    # 로그 스케일은 다음과 같은 상황에서 유용합니다:
        # 데이터의 범위가 매우 넓을 때:
    # 데이터 값의 범위가 매우 클 경우, 로그 스케일을 사용하면 작은 값과 큰 값을 동시에 시각화하기가 더 쉽습니다. 
    # 예를 들어, 1에서 1,000,000까지의 값을 선형 스케일로 표시하면 작은 값들이 거의 보이지 않을 수 있습니다.
        # 지수적 성장 또는 감소를 나타낼 때:
    # 로그 스케일은 지수적 성장(또는 감소)을 선형적으로 표현할 수 있어 이러한 패턴을 더 쉽게 식별할 수 있습니다.


    # fig.show():

# Figure를 화면에 출력합니다.
```

```python
#위 자료 figure에서 위치 바꾸기(배열 인덱스 식)

fig, axs = plt.subplots(2,2)
# ax로 쓰나 axs로 쓰나 상관 없긴 함. 단지 가독성 위해 ㅋㅋ.

axs[0,0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)

axs[1,1].hist(ns_book7['대출건수'], bins=100)
axs[1,1].set_yscale('log')

fig.show()
```

```python
fig, axs = plt.subplots(2, figsize=(6, 8))

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
axs[0].set_title('scatter plot')

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')

fig.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
axs[0].set_title('scatter plot')
axs[0].set_xlabel('number of books')
axs[0].set_ylabel('borrow count')

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')
axs[1].set_xlabel('borrow count')
axs[1].set_ylabel('frequency')

fig.show()
```
